```python
traindf=pd.DataFrame(df['SENTENCES'])
```

```python
traindf.head(5)
traindf.columns
```

```python
 traindf=traindf.rename(index=str, columns={"SENTENCES": "input"})
```

```python
import string
import re
punctuations=string.punctuation
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text
traindf['input']=traindf['input'].apply(lambda x:remove_punct(x))
traindf.head(10)
```

```python
import nltk
traindf['tokenized_input']=traindf.apply(lambda row:nltk.word_tokenize(row['input']),axis=1 )
df.head()
```

```python
traindf['tokenized_input'].head(1)
```

```python
ps = nltk.PorterStemmer()
def text_stemmer(text):
    text = [ps.stem(word) for word in text]
    return text

```

```python
traindf['tokenized_input']=traindf['tokenized_input'].apply(lambda x: text_stemmer(x))
```

```python
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
    text = [word for word in text if word not in stopword]
    return text
traindf['tokenized_input']=traindf['tokenized_input'].apply(lambda x: remove_stopwords(x))
traindf.head(3)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(
            sublinear_tf=True,
            analyzer='word',
            token_pattern=r'\w{1,}',
            ngram_range=(1, 2),
            min_df=5)

overview_text = vectorizer.fit_transform(df['SENTENCES'].fillna(''))
```

```python
print(overview_text.shape)
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
text=TruncatedSVD(n_components=100).fit_transform(overview_text)
```

```python
text.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=20, random_state=20).fit(text)
```
